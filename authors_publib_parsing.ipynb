```python
import requests
from bs4 import BeautifulSoup as BS
import time
import pandas as pd
```

```python
url = 'https://publ.lib.ru/ARCHIVES/_CLASSES/OBS_HUD/_Obs_hud.html'
url_lst = []

page = requests.get(url)
soup = BS(page.content, "html.parser")
divs = soup.find_all('div', class_=['aut2div2', 'aut2div7'])

for div in divs:
    links = div.find_all('a')
    for j in links:
        href = j.get('href')
        text = j.text.strip()
        if href and len(text) > 1:
            full_link = 'https://publ.lib.ru/ARCHIVES' + href[5:]
            try:
                response = requests.get(full_link)
                if response.status_code == 200:
                    url_lst.append(full_link)
            except requests.RequestException:
                continue
url_lst
```

```python
authors = []
headers = {
    'User -Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
for url in url_lst:
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BS(response.content, "html.parser")

        h2_tags = soup.find_all('h2')

        for h2 in h2_tags:
            a_tag = h2.find('a')
            if a_tag and 'href' in a_tag.attrs:
                for nested_tag in a_tag.find_all(class_='fsize1s'):
                    nested_tag.decompose()

                link_text = a_tag.get_text(strip=True)
                authors.append(link_text)
            else:
                print("Вложенный тег <a> не найден в <h2>.")

    except requests.RequestException as e:
        print(f"Ошибка при запросе: {e}")
authors
```

```python
results = []
headers = {
    'User -Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
for full_link in url_lst:
    try:
        response = requests.get(full_link)
        response.raise_for_status()
        link_soup = BS(response.content, "html.parser")
        
        aut1div1 = link_soup.find('div', class_='aut1div1')
        td_elements = link_soup.find_all('td', class_='aut1td2')

        for td in td_elements:
            h4_tag = td.find('h4')
            if h4_tag:
                h4_text = h4_tag.get_text(strip=True)
                if aut1div1:
                    aut1div1_text = aut1div1.get_text(strip=True)
                    results.append([h4_text[1:-1], full_link, aut1div1_text])
    except requests.RequestException as e:
        print(f'Ошибка при запросе к {full_link}: {e}')
    time.sleep(1)

results
```

```python
data = {}
for i in range(len(authors)):
    data[authors[i]] = results[i]
data
```
